#Use pivote table for region wise average GDP,population
A pivot table means:

# “Grouping data based on some category and then applying calculations like mean, sum, count, etc.”

```python
pivot = df.pivot_table(
    values=[
        "GDP: Gross domestic product (million current US$)",
        "Population in thousands (2017)"
    ],
    index="Region",
    aggfunc="mean"
)

pivot.columns = ["Avg GDP", "Avg Population"]
print(pivot)
```

We group countries based on their region (like Asia, Europe), and then calculate the average GDP and population for each group.”
