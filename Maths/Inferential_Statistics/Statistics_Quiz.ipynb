```python
#7) If the mean value is 100 and standard deviation is 15 for a normally distributed sample.
#Calculate the p-value from the given sample after performing the one sample z-test on the following data.
```

```python
from scipy import stats
import numpy as np

Sample = [88, 92, 84, 64, 96, 97, 67, 97, 96, 99, 115, 119, 119, 129, 130, 142, 102, 123, 134, 105]

sample_mean = np.mean(Sample)   # correct
pop_mean = 100                  # given
pop_std_dev = 15
sample_size = len(Sample)

z_stat = (sample_mean - pop_mean) / (pop_std_dev / np.sqrt(sample_size))

pvalue = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # two-tailed

print("Z-stat =", z_stat)
print("p-value =", pvalue)
```

Q: In a town, P(Rain)=1/3.
P(Traffic|Rain)=1/2, P(Traffic|No Rain)=1/4.
P(Late|Rain & Traffic)=1/2,
P(Late|No Rain & No Traffic)=1/8,
Other cases: P(Late)=0.25.

Find:
(a) P(No Rain ∩ Traffic ∩ Not Late)
(b) P(Late)
(c) P(Rain | Late)

A:

(a)
P(No Rain) = 2/3
P(Traffic | No Rain) = 1/4
P(Not Late | No Rain & Traffic) = 1 − 0.25 = 0.75

P = (2/3) × (1/4) × (3/4) = 1/8

---

(b) Total probability of Late:

Rain & Traffic: (1/3)(1/2)(1/2) = 1/12
Rain & No Traffic: (1/3)(1/2)(0.25) = 1/24
No Rain & Traffic: (2/3)(1/4)(0.25) = 1/24
No Rain & No Traffic: (2/3)(3/4)(1/8) = 1/16

Total:
P(Late) = 1/12 + 1/24 + 1/24 + 1/16 = 5/24

---

(c)
P(Rain | Late) = P(Rain ∩ Late) / P(Late)

P(Rain ∩ Late) = 1/12 + 1/24 = 1/8

So,
P(Rain | Late) = (1/8) / (5/24) = 3/5
