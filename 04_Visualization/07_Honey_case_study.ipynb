#1 Which state is having highest honey production,repersent using bar graph
m=df.groupby("state")["production"].max()
max_production=m.max()
state_with_max_production = m.idxmax()
print(max_production, state_with_max_production)

```python
#By drawing bargraph comparing
plt.figure(figsize=(8,5))
sns.barplot(x=[state_with_max_production],y=[max_production],palette='viridis')
plt.title("State with higest honey production")
plt.xlabel("state")
plt.ylabel("Production(units)")
plt.show()
```

```python
num = df.loc[df['state'] == "Alabama"]
print(num)
```

```python
m=df.groupby("state")["production"].max().sort_values(ascending=False).reset_index() #TO CREATE DATAFRAME
m
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=m,y="state", x="production", palette='hsv')
plt.ylabel("State")
plt.xlabel("Production (units)")
plt.title("MAXIMUM HONEY PRODUCTION BY STATE")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

California is the MAXIMUM HONEY PRODUCTION BY STATE


```python
# 2 Which state having Higest and Lowest yeildby colones
yeild=df.groupby('state')["yield_per_colony"].max().sort_values(ascending=False).reset_index()
yeild
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(data=yeild,x='yield_per_colony',y='state',palette="hsv")
plt.show()
```

#Which year witness higest profit in honey production(Bar grap)


```python
profit=df.groupby('year')['value_of_production'].max().sort_values(ascending=False).reset_index()
profit
```

```python
plt.figure(figsize=(20,6))
sns.barplot(data=profit,y="value_of_production",x="year",palette='plasma')
plt.show()
```

#which state having a higest stock in 2020

```python
y1 = df[df['year'] == 2020]
y1.sort_values(by='stocks',ascending=False)
```

```python
plt.figure(figsize=(20,6))
sns.barplot(data=y1, y="state", x="stocks",palette='viridis')
plt.title("Year vs Stocks in Honey Production (for 2020)")
plt.show()
```

# Which state is having higest average price at 21stcenturay

```python
price_21cen=df[df['year']>=2000]
price_21cen.sort_values(by='average_price',ascending=False)
```

```python
df1=price_21cen.groupby('state')['average_price'].mean()
df1
```
