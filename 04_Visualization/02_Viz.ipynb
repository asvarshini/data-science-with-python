```python
#scatter
a=[10,20,30,40]
b=[22,44,55,11]
c=[1,2,3,4]
```

```python
import matplotlib.pyplot as plt
plt.scatter(a,c)
```

```python
#To remove <matplotlib.collections.PathCollection at 0x794fe566fec0>
plt.scatter(a,c)
plt.show()
```

```python
#to mention x,yaxis and label
plt.scatter(a,c)
plt.xlabel(c)
plt.ylabel(a)
plt.title("Satter graph")
plt.show()
```

```python
#if u want boxes
plt.scatter(a,c)
plt.xlabel(a)
plt.ylabel(c)
plt.title("Scatter graph")
plt.grid()
plt.show()
```

```python
#if u want to change colour
plt.scatter(a,c,color ="pink")
plt.xlabel(a)
plt.ylabel(c)
plt.title("Scatter graph")
plt.grid()
plt.show()
```

```python
#if u want to change colour
plt.scatter(a,c,color ="yellow",marker="d")
plt.scatter(b,c,color="red",marker="*")
plt.xlabel(a)
plt.ylabel(c)
plt.title("Scatter graph")
plt.grid()
plt.show()
```

```python
#you want to label what shapr you are using
plt.scatter(a,c,color ="yellow",marker="x",label='person A')
plt.scatter(b,c,color="red",marker="*",label='person B')
plt.xlabel(a)
plt.ylabel(c)
plt.title("Scatter graph")
plt.grid()
plt.legend()
plt.show()
```

```python
#subgraph
plt.subplot(1,2,1)  #1,row 2 coloum 1st graph
plt.scatter(c,a)
plt.ylabel("Products")
plt.subplot(1,2,2)
plt.xlabel("months")
plt.scatter(c,b)
plt.show()
```
