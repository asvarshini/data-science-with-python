```python
import matplotlib.pyplot as plt
A=[12,21,3,55]
B=[22,44,11,33]
C=[1,2,3,4]
```

```python
fig ,axs=plt.subplots(2,2)
axs[0,0].scatter(C,A)
axs[0,0].set_ylabel("product")
axs[0,1].plot(C,A)
axs[1,0].scatter(C,B)
axs[1,1].plot(C,B)
plt.show()
```

```python
fig.savefig("plots .pdf")
```
