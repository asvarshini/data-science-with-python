```python
import pandas as pd
import numpy as np

data = {
    "Employee_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Arun", "Meena", "John", "Priya", "Ravi", "Sara"],
    "Department": ["HR", "IT", "IT", "Finance", None, "HR"],
    "Age": [25, 28, None, 32, 30, None],
    "Salary": [30000, 45000, 40000, None, 38000, 42000],
    "Experience_Years": [2, 5, 3, 7, None, 4]
}

df = pd.DataFrame(data)

print(df)
```

```python
#2.seaborn  3)Ploty 4)power bi
import seaborn as sns
sns.histplot(data["Salary"])
```

```python
#5 types of important charts
#1)Line chart(Best for treand over time)
import matplotlib.pyplot as plt
days =[1,2,3]
sales=[100,200,300]
plt.plot(days,sales)
plt.xlabel(days)
plt.ylabel(sales)
plt.title("Sales Trend")
plt.show()
```

```python
#Example for line chart
import matplotlib.pyplot as plt
Day=["MONDAY","TUESDAY","WEDNESDAY","THURSDAY","FRIDAY","SATURDAY","SUNDAY"]
Temp=[22,33,44,55,54,34,36]
plt.plot(Day,Temp)
plt.title("WEEK TEMPERATURE")
plt.show()
```

```python
#Bar(Comparision)
import matplotlib.pyplot as plt
dep=["HR","IT","FINACE"]
emp=[10,50,20]
plt.bar(dep,emp)
plt.titlem("Company")
plt.show()
```

```python
#Example for bar-chart
import matplotlib.pyplot as plt
Content =["Comedy","Action","Drama"]
View=["100k","300k","700k"]
plt.bar(Content,View)
plt.title("Netflix show")
plt.show()
```

```python
#3)Histogram(Distribution of numercial data)
import matplotlib.pyplot as plt
x=[s for s in data["Salary"] if s is not None]

plt.hist(x,bins=10)
plt.show()
```

```python
# Example for Histograph
import matplotlib.pyplot as plt
x=[10,12,13,14,20,22,25,30,35]
plt.hist(x,bins=10)
plt.show()
```

```python
#4) scatter graph(Relation between 2 data)
import matplotlib.pyplot as plt
hour=[2,3,4,5]
markes=[88,90,99,100]
plt.scatter(hour,markes)
plt.show()
```

```python
#Box plot(To find median,quartiles,outliers)
import matplotlib.pyplot as sns
x=[s for s in data["Salary"] if s is not None]

sns.boxplot(x)
sns.show()
```

```python
#Heatmap (corelation between variable)
import seaborn as sns
sns.heatmap(df.corr(numeric_only=True), annot=True)
```
