```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Cars (1).csv", skipinitialspace=True)
df
```

```python
df.info()
```

```python
#but they are asking with respect to brand then use seaborn
fig,axs=plt.subplots(2,2,figsize=(20,10))
plt.subplots_adjust(wspace=0.5,hspace=0.5)

sns.scatterplot(x=df["mpg"], y=df["hp"], hue=df["brand"], ax=axs[0,0])
axs[0,0].set_xlabel("mpg")
axs[0,0].set_ylabel("hp")

sns.scatterplot(x=df["mpg"], y=df['weightlbs'], hue=df["brand"], ax=axs[0,1])
axs[0,1].set_xlabel("mpg")
axs[0,1].set_ylabel("weight")

sns.scatterplot(x=df["hp"], y=df["mpg"], hue=df["brand"], ax=axs[1,0])
axs[1,0].set_xlabel("hp")
axs[1,0].set_ylabel("mpg") # Corrected from 'weight' to 'mpg'

sns.scatterplot(x=df["hp"], y=df["time-to-60"], hue=df["brand"], ax=axs[1,1])
axs[1,1].set_xlabel("hp")
axs[1,1].set_ylabel("time-to-60")

plt.show()
```

```python
fig,axs=plt.subplots(2,2,figsize=(20,10))
plt.subplots_adjust(wspace=0.5,hspace=0.5)

sns.barplot(x="cylinders", y="hp", hue="brand", data=df, ax=axs[0,0])
axs[0,0].set_xlabel("Cylinders")
axs[0,0].set_ylabel("hp")

sns.barplot(x="cylinders", y="mpg", hue="brand", data=df, ax=axs[0,1])
axs[0,1].set_xlabel("Cylinders")
axs[0,1].set_ylabel("mpg")

sns.barplot(x="cylinders", y="weightlbs", hue="brand", data=df, ax=axs[1,0])

axs[1,0].set_xlabel("Cylinders")
axs[1,0].set_ylabel("Weight (lbs)")
axs[1,0].set_title("Car Weight by Cylinders and Brand")

sns.barplot(x="cylinders", y="time-to-60", hue="brand", data=df, ax=axs[1,1])
axs[1,1].set_xlabel("Cylinders")
axs[1,1].set_ylabel("Time to 60")

plt.show()
#showing higest cylinders higest horse power
```

box plot

```python
fig,axs=plt.subplots(2,2,figsize=(20,10))
plt.subplots_adjust(wspace=0.5,hspace=0.5)

sns.boxplot(x="cylinders", y="hp", hue="brand", data=df, ax=axs[0,0])
axs[0,0].set_xlabel("Cylinders")
axs[0,0].set_ylabel("hp")

sns.boxplot(x="cylinders", y="mpg", hue="brand", data=df, ax=axs[0,1])
axs[0,1].set_xlabel("Cylinders")
axs[0,1].set_ylabel("mpg")

sns.boxplot(x="cylinders", y="weightlbs", hue="brand", data=df, ax=axs[1,0])
axs[1,0].set_xlabel("Cylinders")
axs[1,0].set_ylabel('Weight (lbs)')

sns.boxplot(x="cylinders", y="time-to-60", hue="brand", data=df, ax=axs[1,1])
axs[1,1].set_xlabel("Cylinders")
axs[1,1].set_ylabel("Time to 60")

plt.show()
#showing higest cylinders higest horse power
```

Create histograph,hp,mpg,time_to_60,weight  wrt to brand.All should present in same plane

```python
fig,axs=plt.subplots(2,2,figsize=(20,10))
plt.subplots_adjust(wspace=0.5,hspace=0.5)

sns.histplot(x="cylinders", hue="brand", data=df, ax=axs[0,0])
axs[0,0].set_xlabel("Cylinders")
axs[0,0].set_ylabel("Count") # Default for histplot is count on y-axis, if y is not specified

sns.histplot(x="mpg", hue="brand", data=df, ax=axs[0,1])
axs[0,1].set_xlabel("mpg")
axs[0,1].set_ylabel("Count")

sns.histplot(x="weightlbs", hue="brand", data=df, ax=axs[1,0])
axs[1,0].set_xlabel("weightlbs")
axs[1,0].set_ylabel('Count')

sns.histplot(x="time-to-60", hue="brand", data=df, ax=axs[1,1])
axs[1,1].set_xlabel("time-to-60")
axs[1,1].set_ylabel("Count")

#it showing range like between 10 t0 15 mpg range20 cars were present
```
