```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/car.csv")
df
```

```python
df.info()
```

#Create scatter plot for the following data and make sure all plots apper in thr same frame
#(mpg-hp,mpg-weight,hp-mpg,hp-timetoweight)with respect to brand

```python
fig,axs=plt.subplots(2,2)
plt.subplots_adjust(wspace=0.5,hspace=0.5)

axs[0,0].scatter(df["mpg"],df["HP"])
axs[0,0].set_xlabel("mpg")
axs[0,0].set_ylabel("HP")
axs[0,1].scatter(df["mpg"],df['weight_lbs'])
axs[0,1].set_xlabel("mpg")
axs[0,1].set_ylabel("weight")
axs[1,0].scatter(df["HP"],df["mpg"])
axs[1,0].set_xlabel("HP")
axs[1,0].set_ylabel("weight")
axs[1,1].scatter(df["HP"],df["time_to_60"])
axs[1,1].set_xlabel("HP")
axs[1,1].set_ylabel("time_to_60")
plt.show()
```

```python
#but they are asking with respect to brand then use seaborn
fig,axs=plt.subplots(2,2,figsize=(20,10))
plt.subplots_adjust(wspace=0.5,hspace=0.5)

sns.scatterplot(x=df["mpg"], y=df["HP"], hue=df["brand"], ax=axs[0,0])
axs[0,0].set_xlabel("mpg")
axs[0,0].set_ylabel("HP")

sns.scatterplot(x=df["mpg"], y=df['weight_lbs'], hue=df["brand"], ax=axs[0,1])
axs[0,1].set_xlabel("mpg")
axs[0,1].set_ylabel("weight")

sns.scatterplot(x=df["HP"], y=df["mpg"], hue=df["brand"], ax=axs[1,0])
axs[1,0].set_xlabel("HP")
axs[1,0].set_ylabel("mpg") # Corrected from 'weight' to 'mpg'

sns.scatterplot(x=df["HP"], y=df["time_to_60"], hue=df["brand"], ax=axs[1,1])
axs[1,1].set_xlabel("HP")
axs[1,1].set_ylabel("time_to_60")

plt.show()
```

#create barplot ,hp,mpg,weight,time_to_60 with r to brand and no of cylinders


```python
fig,axs=plt.subplots(2,2,figsize=(20,10))
plt.subplots_adjust(wspace=0.5,hspace=0.5)

sns.barplot(x="cylinders", y="HP", hue="brand", data=df, ax=axs[0,0])
axs[0,0].set_xlabel("Cylinders")
axs[0,0].set_ylabel("HP")

sns.barplot(x="cylinders", y="mpg", hue="brand", data=df, ax=axs[0,1])
axs[0,1].set_xlabel("Cylinders")
axs[0,1].set_ylabel("mpg")

sns.barplot(x="cylinders", y="weight_lbs", hue="brand", data=df, ax=axs[1,0])
axs[1,0].set_xlabel("Cylinders")
axs[1,0].set_ylabel('Weight (lbs)')

sns.barplot(x="cylinders", y="time_to_60", hue="brand", data=df, ax=axs[1,1])
axs[1,1].set_xlabel("Cylinders")
axs[1,1].set_ylabel("Time to 60")

plt.show()
```
