```python
import pandas as pd
import numpy as np

data = {
    "Employee_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Arun", "Meena", "John", "Priya", "Ravi", "Sara"],
    "Department": ["HR", "IT", "IT", "Finance", "HR", "HR"],
    "Age": [25, 28, 21, 32, 30, 21],
    "Salary": [30000, 45000, 40000,2000, 38000, 42000],
    "Experience_Years": [2, 5, 3, 7, 2, 4]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.loc[df['Age']>25, 'Name'])
```

```python
print(sorted(df["Age"]))
```

```python
#If u want to sort age with name also then
df = df.sort_values(by='Age')
df
```

```python
#Using for loop print ge and and store in set
se = set()
lis=[]
for i in df['Age']:
  print(i)
  se.add(i)
  lis.append(i)
print(se)
print(lis)
```
