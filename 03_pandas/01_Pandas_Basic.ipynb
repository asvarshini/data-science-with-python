```python
#Filtering Data
import pandas as pd

data = {
    "Name": ["Rahul", "Sneha", "Arjun", "Meera", "Kiran", "Anjali"],
    "Age": [25, 28, 24, 30, 27, 26],
    "Salary": [35000, 42000, 30000, 50000, 38000, 41000],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Finance"]
}

df = pd.DataFrame(data)

print(df)
```

```python

df[df["Age"]>25]
```

```python
#fliterning more than one Data
df[(df["Age"]>23) & (df["Salary"]>38000)]
```

```python
#Adding new coloum
df["Bonus"]=df["Salary"]*0.10
print(df)
```

```python
import pandas as pd
import numpy as np

data = {
    "Employee_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Arun", "Meena", "John", "Priya", "Ravi", "Sara"],
    "Department": ["HR", "IT", "IT", "Finance", None, "HR"],
    "Age": [25, 28, None, 32, 30, None],
    "Salary": [30000, 45000, 40000, None, 38000, 42000],
    "Experience_Years": [2, 5, 3, 7, None, 4]
}

df = pd.DataFrame(data)

print(df)
```

```python
#handling missing values
#is null if value is missing True or False(if value is present)

df.isnull()
```

```python
#df.isnull().sum()
#it will give sum of null value in coloum
df.isnull().sum()
```

```python
#df.dropna()
#remove the rows contains missing value
df.dropna()
```

```python
#df.fillna()
df.fillna("a")
```

```python
df.describe()
```

```python
#by using fillna i can also fill mean and median value
df["Age"].fillna(df["Age"].mean())
```

```python
# Fill Median
df["Salary"].fillna(df["Salary"].median())
```

```python
#forward fillng
df.fillna(method="ffill")
```

```python
#backward filling
df.fillna(method="bfill")
```

```python
#group by
df.groupby("Age")["Salary"].mean()
```
