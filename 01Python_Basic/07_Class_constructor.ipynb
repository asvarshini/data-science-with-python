```python
#calculator code
class  calculator:
    def __init__(self,a,b):
      self.a =a
      self.b=b
    def add(self):
      return self.a+self.b
s1=calculator(10,20)
print(s1.add())
```

```python
class interset:
  def __init__(self,p,t,r):
    self.p =p
    self.t=t
    self.r=r
  def simple(self):
    return (self.p*self.t*self.r)/100
  def comp(self):
    return (self.p*(1+(self.r/100))**self.t)-self.p
p =int(input("enter the value of p"))
t=int(input("enter the value t"))
r =int(input("ente rthe value of r"))
ob=interset(p,t,r)
print(ob.simple())
print(ob.comp())
```
