```python
#Find common elements between two lists.
li1 =[2,4,67]
li2=[4,2,6,5]
common=[]

# A correct way to find common elements is to iterate through one list
# and check for presence in the other list. Using sets is often more efficient.
# Method 1: Using loops
# for element_li1 in li1:
#   if element_li1 in li2 and element_li1 not in common:
#     common.append(element_li1)

# Method 2: Using sets (more Pythonic and efficient for larger lists)
common = list(set(li1) & set(li2))

print("common element",common)
```

```python
# Print all prime numbers between 1 and N.
n = int(input("Enter the number")) # Corrected typo 'nubmer' to 'number'
for num in range(2, n + 1):
  is_prime = True
  # Check for factors from 2 up to the square root of num
  # Convert the upper bound of range to an integer
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      is_prime = False
      break # If a factor is found, it's not prime, so break

  # The original line 'is_prime = True' here was a logical error.
  # It would reset is_prime to True for every number, even if it was set to False.
  # Now, we only print if is_prime remains True after the checks.
  if is_prime:
    print(num)
```

```python
#Find missing number in list from 1 to N.
List = [1, 2, 4, 5]
def missing(List):
  n =len(List)+1
  actual_sum =n*(n+1)//2
  excpeted_sum =sum(List)
  return actual_sum -excpeted_sum
print(missing(List))
```
