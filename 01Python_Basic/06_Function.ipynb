```python
# Function is A function is a reusable block of code.
def greet():
    print("Hello")
```

```python
#Parameters & Return
def add(a, b):
    return a + b
result = add(2,3)
print(result)
```

```python
#Default Arguments
def greet(name="Guest"):
    print(name)
greet()
greet("Varshini")
```

```python
#*args (Multiple Positional Arguments)
def total(*numbers):
    print(numbers)
total(1,2,3,4)
```

```python
def info(**data):
    print(data)
info(name="Varshini", age=22)
```

```python
#LAMBDA FUNCTION
#normal
def square(x):
    return x*x
print(square(3))

```

```python
#lambda
square = lambda x: x*x
print(square(5))
```

```python
#Mulitiple arguments
add =lambda x,y:x+y
print(add(5,5))
```

```python
check_even = lambda x: "Even" if x%2==0 else "Odd"

print(check_even(10))
```

```python
# with lambda
lis =[1,2,3]
sq =list(map(lambda x:x*x,lis))
print(sq)
```

```python
#with fliter
lis =[1,23,2,3,4]
num=list(filter(lambda x :x%2==0,lis))
print(num)
```

```python
#with sorted
students = [("A",90),("B",70),("C",85)]
ex=sorted(students, key=lambda x:x[1])
print(ex)
```

```python
#Recursion
def fact(n):
  if n ==0:
    return 1

  return n*fact(n-1)
print(fact(5))
```

```python
#Expection hadling
def divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Cannot divide by zero"
print(divide(10,5))
print(divide(10,0))

```

```python
def read_file(filename):
    try:
        with open(filename, "r") as f:
            return f.read()
    except FileNotFoundError:
        return "File not found"
print(read_file("ammu"))
```
