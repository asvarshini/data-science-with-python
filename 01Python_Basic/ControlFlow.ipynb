```python
#if-elif-else ladder
marks = 85

if marks >= 90:
    print("A")
elif marks >= 75:
    print("B")
elif marks >= 50:
    print("C")
else:
    print("Fail")
```

```python
#nested if
age = 20
if age >= 18:
    if age < 60:
        print("Adult")
```
