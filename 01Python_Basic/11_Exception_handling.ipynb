```python
#Inside function,without return function
def ex(a,b):
  try:
    print("re=",a/b)
  except ZeroDivisionError:
    print("Cannot division")
ex(10,5)
ex(10,0)
```

```python
#Inside function,without return function

def ex(a,b):
  try:
    return "re=",a/b
  except ZeroDivisionError:
    return("Cannot division")
print(ex(10,5))
print(ex(10,0))
```

```python
#Multipal ExpectionHandling
def calculate(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Zero division error"
    except TypeError:
        return "Invalid datatype"
calculate(10,"2")
```

```python
#File ExpectionHandling
try:
  f=open("data.txt","r")
  content=f.read()
  print(content)
  f.close()
except FileNotFoundError:
  print("File not found")
```

```python
import os

# Check current directory
print("Current directory:", os.getcwd())

def read_file(filepath):
    try:
        with open(filepath, "r") as f:
            return f.read()
    except FileNotFoundError:
        return "File Not Found"
    except PermissionError:
        return "Permission Denied"
    except Exception as e:
        return f"Unexpected error: {e}"

# Call the function
print(read_file("overriding.ipynb"))
```
