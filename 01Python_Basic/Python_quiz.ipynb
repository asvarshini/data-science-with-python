```python
class Parent:
  def __init__(self):

    self._name = "Levi"

    self.__age = 29

class Child(Parent):

  def __init__(self):

    Parent.__init__(self)

    self._name = "Devon"

    self.__age = 28

  def view(self):

    return self.__age

obj = Child()

obj.view()
```
