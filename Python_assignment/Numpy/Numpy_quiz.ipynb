# 9.For the given arrays, array1 and array2, if we stack the two arrays column wise, what will
be the output?

```python
#When we stack arrays column-wise, we are placing one array as a new column next to another array.
array1 = np.array([[1,2],[3,4]])
array2 = np.array([[5,6],[7,8]])
print(np.column_stack((array1, array2)))
```
